```python
def convert_np_to_python(obj):
    if isinstance(obj, np.ndarray):
        return obj.tolist()
    elif isinstance(obj, (np.float32, np.float64)):
        return float(obj)
    elif isinstance(obj, dict):  # Jeśli to słownik, konwertujemy wartości
        return {k: convert_np_to_python(v) for k, v in obj.items()}
    return obj
```

```python
results_python = convert_np_to_python(results)
```
