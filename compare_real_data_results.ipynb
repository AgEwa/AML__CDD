```python
import json

import numpy as np

from real_data_experiments import compare_betas

np.random.seed(42)
```

```python
for dataname in ['gisette', 'bioresponse', 'secom', 'sonar']:
    print(f"\nComparison for {dataname}")
    with open(f'results/{dataname}.json', 'r') as file:
        result = json.load(file)
    for metric in result.keys():
        print(f'{metric} comparison:\nCDD: {result[metric]["score_ccd"]}, logreg: {result[metric]["score_log_reg"]}')
        distance = compare_betas(result[metric]["best_beta"], result[metric]["best_beta_log_reg"])
        print(f"Mean euclidean distance of betas: {distance}")
```
